# 01 - Ver radiografías (manual)
Sube imágenes a `data/raw/manual/` y ejecútalo.


```python
import glob
from PIL import Image
import matplotlib.pyplot as plt

paths = sorted(glob.glob('data/raw/manual/*'))
paths = [p for p in paths if p.lower().endswith(('.png','.jpg','.jpeg'))]
print('Imágenes encontradas:', len(paths))
paths[:10]
```

```python
n = min(12, len(paths))
cols = 4
rows = (n + cols - 1)//cols
plt.figure(figsize=(12, 3*rows))
for i,p in enumerate(paths[:n]):
    img = Image.open(p).convert('L')
    plt.subplot(rows, cols, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(p.split('/')[-1])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
